```python
#Imports
import os
import seaborn as sns
import numpy as np
import nltk
import pandas as pd
import matplotlib.pyplot as plt
from tika import parser
from nltk.corpus import stopwords
from string import digits
from sklearn.feature_extraction import text
from sklearn import preprocessing
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import HashingVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import TruncatedSVD
from yellowbrick.cluster import KElbowVisualizer
from yellowbrick.cluster import SilhouetteVisualizer
from yellowbrick.cluster import InterclusterDistance
from wordcloud import WordCloud
```

```python
#Loading data
BASE_LOCATION = 'C:\\Users\\ifta\\papers\\'
pdfs = os.listdir(BASE_LOCATION)
paper_contents=[]
paper_titles=[]
count = 1
s=set(stopwords.words('english'))
s.add('Stack')
s.add('Overflow')
s.add('pdf')
remove_digits = str.maketrans('', '', digits)
for paper in pdfs:
    title = paper.rsplit('.',1)[0]
    print('[',count,'/',len(pdfs),'] Parsing paper: ',title)
    raw = parser.from_file(BASE_LOCATION+paper)
    paper_content = raw['content']
    tokens = paper_content.split()
    tokens = list(filter(lambda w: not w in s and w.isalpha() and len(w)>2,tokens))
    paper_contents.append(" ".join(tokens))
    paper_titles.append(title)
    count+=1
print("Loaded titles = ",len(paper_titles))
print("Loaded contents = ",len(paper_contents))
```

```python
stopwords = nltk.corpus.stopwords.words('english')
for i in text.ENGLISH_STOP_WORDS:
    stopwords.append(i)
#print(stopwords)
#vectorizer = CountVectorizer()
vectorizer = TfidfVectorizer(min_df=0.8,stop_words=stopwords)
#hasher = HashingVectorizer(n_features=120, stop_words=stopwords, norm='l2', binary=False)
#vectorizer = make_pipeline(hasher, TfidfTransformer())
X = vectorizer.fit_transform(paper_contents)
```

```python
print(X.shape)
```

```python
#Finding the elbow
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,15), timings=False)
visualizer.fit(X)
visualizer.show()
```

```python
#Clustering
true_k = 6
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=200, n_init=10)
model.fit(X)
labels = model.labels_
paper_clusters = pd.DataFrame(list(zip(paper_titles,labels)),columns=['title','cluster'])
print(paper_clusters.sort_values(by=['cluster']))
```

```python
#Visualization on a web page
webpage = open("result.html","w")
def add_to_webpage(data):
    webpage.write(data)
def close_webpage():
    add_to_webpage('</body></html>\n')
    webpage.flush()
    webpage.close()

def new_line_to_webpage():
    webpage.write('<br>\n')
    
add_to_webpage('<!DOCTYPE HTML>\n<html>\n')
add_to_webpage('<head><meta charset="utf-8"><br><title>Stack Overflow Research Document Clustering</title><br><link rel=\'stylesheet\' href=\'style.css\' type=\'text/css\' /></head>\n')
add_to_webpage('<body>\n')

def wordcloud():
    result={'cluster':labels,'content':paper_contents}
    result=pd.DataFrame(result)
    add_to_webpage('<h1>Cluster details</h1>')
    for k in range(0,true_k):
        s=result[result.cluster==k]
        text=s['content'].str.cat(sep=' ')
        text=text.lower()
        text=' '.join([word for word in text.split()])
        wordcloud = WordCloud(max_font_size=50, max_words=30, background_color="white").generate(text)
        clusterNumber = 'Cluster {}'.format(k)
        print(clusterNumber)
        add_to_webpage('<b>{}</b>'.format(clusterNumber))
        new_line_to_webpage()
        add_to_webpage('<b>Titles</b>')
        new_line_to_webpage()
        #print('Titles')
        titles=paper_clusters[paper_clusters.cluster==k]['title']
        count = 1
        for t in titles:
            add_to_webpage('{0}. {1}'.format(count,t))
            new_line_to_webpage()
            count+=1
        #print(titles.to_string(index=False))
        plt.figure()
        plt.imshow(wordcloud, interpolation="bilinear")
        plt.axis("off")
        #plt.show()
        imageName = clusterNumber+'.png'
        plt.savefig(imageName)
        imageData = '<img src=\'{}\'/>'.format(imageName)
        add_to_webpage(imageData)
        new_line_to_webpage()
        
def cluster_view():
    add_to_webpage('<h1>Cluster visualization</h1>')
    clf = TruncatedSVD(X.shape[1]-1)
    df = clf.fit_transform(X)
    unique_labels = np.unique(labels)
 
    for i in unique_labels:
        plt.scatter(df[labels == i , 0] , df[labels == i , 1] , label = i)
    plt.legend()
    imageName = 'visualization.png'
    plt.savefig(imageName)
    imageData = '<img src=\'{}\' class=\'c-img\'/>'.format(imageName)
    add_to_webpage(imageData)
    new_line_to_webpage()
    
cluster_view()
wordcloud()
close_webpage()
```
